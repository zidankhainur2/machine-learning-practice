```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameter
l = m = 5
dx = dy = 2.0
dt = 0.05
T = 10

# Fungsi awal Φ(x, y) = sech^2(x^2 + y^2)
def phi(x, y):
    return 1 / np.cosh(x**2 + y**2)**2

# Inisialisasi grid
x = np.arange(-5, (l + 1) * dx - 5, dx)
y = np.arange(-5, (m + 1) * dy - 5, dy)
X, Y = np.meshgrid(x, y)

# Inisialisasi solusi u pada waktu awal
u = phi(X, Y)

# Metode beda hingga skema eksplisit CTCS (Leapfrog)
def solve_wave_equation(u, l, m, dx, dy, dt, T):
    num_steps = int(T / dt)

    for step in range(1, num_steps + 1):
        # Hitung nilai u pada waktu selanjutnya menggunakan CTCS
        u_next = np.zeros((l + 1, m + 1))
        for i in range(1, l):
            for j in range(1, m):
                u_next[i, j] = 2 * u[i, j] - u[i, j] + (dt/dx)**2 * (u[i+1, j] - 2*u[i, j] + u[i-1, j]) + \
                               (dt/dy)**2 * (u[i, j+1] - 2*u[i, j] + u[i, j-1])

        # Update nilai u untuk iterasi berikutnya
        u = u_next.copy()

    return u

# Solusi numerik
u_numerical = solve_wave_equation(u, l, m, dx, dy, dt, T)

# Visualisasi
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, u_numerical, cmap='viridis')  # Adjust dimensions
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('u(x, y, t)')
ax.set_title('Solusi Numerik Persamaan Gelombang 2D dengan Metode CTCS')
ax.set_zlim(-2, 5)

plt.show()
```
