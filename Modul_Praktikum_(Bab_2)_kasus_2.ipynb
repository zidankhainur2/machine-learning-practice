```python
import heapq

class Puzzle:
    def __init__(self, board, g, parent=None):
        self.board = board
        self.g = g
        self.parent = parent
        self.h = self.misplaced_tiles()
        self.f = self.g + self.h

    def __lt__(self, other):
        return self.f < other.f

    def misplaced_tiles(self):
        goal = [[2, 8, 3], [1, 6, 4], [7, 0, 5]]
        return sum(1 for i in range(3) for j in range(3) if self.board[i][j] != 0 and self.board[i][j] != goal[i][j])

    def generate_successors(self):
        successors = []
        x, y = next((i, j) for i in range(3) for j in range(3) if self.board[i][j] == 0)
        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if 0 <= nx < 3 and 0 <= ny < 3:
                new_board = [row[:] for row in self.board]
                new_board[x][y], new_board[nx][ny] = new_board[nx][ny], new_board[x][y]
                successors.append(Puzzle(new_board, self.g + 1, self))
        return successors

    def is_goal(self):
        return self.board == [[2, 8, 3], [1, 6, 4], [7, 0, 5]]

def print_solution(puzzle):
    path = []
    while puzzle:
        path.append(puzzle.board)
        puzzle = puzzle.parent
    for board in reversed(path):
        for row in board:
            print(row)
        print()

def best_first_search(start):
    open_list = []
    heapq.heappush(open_list, start)
    closed_list = set()

    while open_list:
        current = heapq.heappop(open_list)

        if current.is_goal():
            print_solution(current)
            return

        closed_list.add(tuple(map(tuple, current.board)))

        for successor in current.generate_successors():
            if tuple(map(tuple, successor.board)) not in closed_list:
                heapq.heappush(open_list, successor)

if __name__ == "__main__":
    start_board = [
        [1, 2, 3],
        [4, 0, 5],
        [7, 6, 8]
    ]
    start_puzzle = Puzzle(start_board, 0)
    best_first_search(start_puzzle)
```
