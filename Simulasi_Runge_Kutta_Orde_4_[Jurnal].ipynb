Codingan di bawah adalah implementasi metode Runge-Kutta orde 4 dalam bahasa pemrograman Python untuk mensimulasikan pergerakan partikel berdasarkan persamaan diferensial yang diberikan.

Judul Jurnal **"Simulasi Gerak Osilasi Model Pegas Bergandeng Menggunakan Metode Runge-Kutta"**

Ini akan menghasilkan dua grafik. Yang pertama akan menunjukkan perubahan posisi partikel seiring waktu, dan yang kedua akan menunjukkan perubahan kecepatan partikel seiring waktu.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def runge_kutta_4(f, x0, v0, t0, tn, h):
    n = int((tn - t0) / h)
    t = np.linspace(t0, tn, n+1)
    x = np.zeros(n+1)
    v = np.zeros(n+1)
    x[0] = x0
    v[0] = v0

    for i in range(n):
        k1x = h * v[i]
        k1v = h * f(x[i], v[i])

        k2x = h * (v[i] + 0.5 * k1v)
        k2v = h * f(x[i] + 0.5 * k1x, v[i] + 0.5 * k1v)

        k3x = h * (v[i] + 0.5 * k2v)
        k3v = h * f(x[i] + 0.5 * k2x, v[i] + 0.5 * k2v)

        k4x = h * (v[i] + k3v)
        k4v = h * f(x[i] + k3x, v[i] + k3v)

        x[i+1] = x[i] + (k1x + 2*k2x + 2*k3x + k4x) / 6
        v[i+1] = v[i] + (k1v + 2*k2v + 2*k3v + k4v) / 6

    return t, x, v
```

```python
# Fungsi persamaan diferensial
def f(x, v):
    k = 1
    m = 1
    return -k/m * x
```

```python
# Parameter awal
x0 = 0.1
v0 = 0
t0 = 0
tn = 10
h = 0.01
```

```python
# Menjalankan metode Runge-Kutta orde 4
t, x, v = runge_kutta_4(f, x0, v0, t0, tn, h)
```

```python
# Plot hasil simulasi
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, x, label='Posisi (x)')
plt.xlabel('Waktu (t)')
plt.ylabel('Posisi (x)')
plt.title('Simulasi Posisi Partikel dengan Metode Runge-Kutta Orde 4')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, v, label='Kecepatan (v)')
plt.xlabel('Waktu (t)')
plt.ylabel('Kecepatan (v)')
plt.title('Simulasi Kecepatan Partikel dengan Metode Runge-Kutta Orde 4')
plt.legend()

plt.tight_layout()
plt.show()
```
